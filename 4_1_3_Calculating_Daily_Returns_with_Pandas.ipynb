# 4_1_3_Calculating_Daily_Returns_with_Pandas

```python
# Use the following price information and the shift(1) function to generate a DataFrame of daily returns:
import pandas as pd
stock_abc = pd.DataFrame({'close' : [11.25, 11.98, 10.74, 11.16, 12.35, 12.87, 13.03, 13.14, 13.37, 12.99]})
```

```python
# View the DataFrame
stock_abc
```

```python
# Create the daily_returns Dataframe from the stock_abc prices using the shift(1) function
daily_returns = (stock_abc - stock_abc.shift(1)) / stock_abc.shift(1)

# View the daily_returns DataFrame
daily_returns
```

```python
# What is the weekly return value from one Monday to the next (11.25 to 12.87), a period of five trading days, for the stock_abc?
# Assume the first value is the closing price for a Monday.
```

```python
# Create the weekly_returns Dataframe from the stock_abc prices using the shift(5) function
weekly_returns = (stock_abc - stock_abc.shift(5)) / stock_abc.shift(5)

# View the daily_returns DataFrame
weekly_returns
```

```python
# Create a new column "Previous Close" for viewing purposes
stock_abc["previous close"] = stock_abc.shift(1)

# View the updated DataFrame
stock_abc
```

```python
# Create a new column "previous week close" for viewing purposes
stock_abc["previous week close"] = stock_abc["close"].shift(5)

# View the updated DataFrame
stock_abc
```

```python
# Use the pct_change function to generate a DataFrame of daily returns.
# Compare the resulting DataFrame to the one that you generated by using shift(1).
# Do the two DataFrames have matching values?

# View the daily_returns DataFrame that was generated using shift(1)
daily_returns
```

```python
# Generate the new daily_returns DataFrame using the pct_change function
daily_returns = stock_abc["close"].pct_change()

# View the new daily_returns DataFrame that was generated using the pct_change function
daily_returns 
```
