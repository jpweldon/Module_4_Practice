# Variance Exercise

```python
# Import the pandas library
import pandas as pd

# Create the stock DataFrame
stocks = pd.DataFrame({
'stock_hij' : [142.13, 143.54, 142.52, 143.43, 141.73, 137.68, 139.71, 139.60, 139.61, 141.57],
'stock_klm' : [53.88, 52.83, 52.55, 53.84, 54.16, 55.58, 53.43, 52.50, 53.90, 47.86],
'market' : [3534.22, 3511.93, 3488.67, 3483.34, 3483.81, 3426.92, 3443.12, 3435.56, 3453.49, 3449.52]
})
```

```python
# Create a daily_returns DataFrame by using the pct_change and dropna functions.
daily_returns = stocks.pct_change().dropna()
```

```python
# Create a default line plot to visualize the volatility of each of the three stocks.
daily_returns.plot(figsize=(15, 10), title='Daily Returns of Stock HIJ, Stock KLM, and the Market', ylabel='Daily Returns')
```

```python
# Use the var function to calculate the variance for each of the three stocks.
# Sorting values with the sort_values function.
daily_returns.var().sort_values()
```
